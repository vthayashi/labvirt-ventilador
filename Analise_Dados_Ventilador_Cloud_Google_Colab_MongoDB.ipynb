## Dataprep function definition

```
def dataprep(velocidade, rede, corrente):

  velocidade_bin = []

  for datapoint in velocidade:
    if datapoint < 0.5:
      velocidade_bin.append(0)
    else:
      velocidade_bin.append(1)


  velocidade_proc_time = []

  state = 0
  prevstate = 0

  time = 0
  prevtime = 0

  for i in range(len(velocidade_bin)):
    time = i
    state = velocidade_bin[i]

    if ((prevstate == 0)and(state == 1)):
      velocidade_proc_time.append(time)

    prevstate = state


    velocidade_proc = []

    prevtime = 0
    time = 0

    for datapoint in velocidade_proc_time:
      time = datapoint
      if (time != prevtime):
        velocidade_proc.append(20/(time - prevtime))
      prevtime = time



  rede_bin = []

  for datapoint in rede:
    if datapoint < 0.5:
      rede_bin.append(0)
    else:
      rede_bin.append(1)


  rede_proc_time = []

  state = 0
  prevstate = 0

  time = 0
  prevtime = 0

  for i in range(len(rede_bin)):
    time = i
    state = rede_bin[i]

    if ((prevstate == 0)and(state == 1)):
      rede_proc_time.append(time)

    prevstate = state

  rede_inicio = rede_proc_time[1]



  corrente_max = 0
  corrente_min = 2.5
  corrente_amp = []
  corrente_amp_time = []

  corrente = corrente[::-1]

  for i in range(len(corrente)):
    datapoint = corrente[i]

    if datapoint < corrente_min:
      corrente_min = datapoint
    if datapoint > corrente_max:
      corrente_max = datapoint

    corrente_amp.append(corrente_max - corrente_min)
    corrente_amp_time.append(i)

  corrente = corrente[::-1]
  corrente_amp = corrente_amp[::-1]

  corrente_amp = corrente_amp[:1900]
  corrente_amp_time = corrente_amp_time[:1900]



  for i in range(len(corrente_amp)):
    datapoint = corrente_amp[i]
    delta = datapoint - corrente_amp[len(corrente_amp)-1]
    
    if (delta < 0.1):
      start_time = corrente_amp_time[i]
      return(start_time,rede_inicio,max(velocidade_proc))
```

```
dataprep(velocidade,rede,corrente)
```

## KNN Clustering Algorithm

### Setup

import libraries

```
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```
experiment_lists = []
```

```
db = client.Ventilador
experiments = db.dados

for experiment in experiments.find({}):

  id = experiment['_id']
  velocidade_anotada = experiment['nome'][len(experiment['nome']) - 1]
  velocidade = experiment['wa'][0]
  rede = experiment['wa'][1]
  corrente = experiment['wa'][2]

  start_time, rede_inicio, velocidade_max = dataprep(velocidade,rede,corrente)

  experiment_list = [start_time, rede_inicio, velocidade_max, id, velocidade_anotada]

  experiment_lists.append(experiment_list)

print(experiment_lists)
```

```
df = pd.DataFrame(experiment_lists, columns=['start','network','velocity', 'id', 'velocidade_anotada'])
```

```
df
```

## Clustering with start and velocidade_anotada

```
df2 = df[['start', 'velocidade_anotada']].copy()

kmeans = KMeans(n_clusters=3).fit(df2)
centroids = kmeans.cluster_centers_

plt.figure(figsize = (8,8))

plt.scatter(df2['start'], df2['velocidade_anotada'],
            c= kmeans.labels_.astype(float), s=50, alpha=0.5)
#plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.xlabel("time to start")
plt.ylabel("labeled velocity")

plt.show()
```
